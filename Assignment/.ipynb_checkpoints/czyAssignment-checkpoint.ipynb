# Filling Missing Values & Data Pre-processing

```python
df2 = df.copy()
df2.head()
```

```python
#Fill in missing values
for col in df2.columns:
    if df2[col].dtype == 'object':
        df2[col] = df2[col].fillna(df2[col].mode()[0])
    else:
        df2[col] = df2[col].fillna(df2[col].median())
```

```python
#Check for missing values
df2.isnull().sum()
```

```python
#Encode 'Disorder Subclass'
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
df2['Disorder Subclass'] = labelEncoder.fit_transform(df2['Disorder Subclass'])
df2.head()
```

```python
num_col
```

```python
#Set all numerical column values into float values
for col in num_col:
    df2[col] = df2[col].astype('float')
```

```python
y1 = df2.pop('Disorder Subclass')
```

```python
#Standardize values in column to be standardized
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df2[col_to_standardized] = scaler.fit_transform(df2[col_to_standardized])
df2.head()
```

```python
# Perform one hot encoding on categorical columns
df2 = pd.get_dummies(data=df2, drop_first=True)
df2
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df2, y1)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
```

# Models
# SVM

```python
from sklearn.svm import SVC
#An SVM model using linear kernel
classifier = SVC(C=1, kernel='linear')

# fit classifier to training set
print(classifier.fit(X_train, y_train))

# make predictions on test set
print(f'Accuracy: {classifier.score(X_test, y_test):.4f}')
```

```python
#An SVM model using Radial Basis Function, or Gaussian Kernel
classifier = SVC(C=1, kernel='rbf')

# fit classifier to training set
print(classifier.fit(X_train, y_train))

# make predictions on test set
print(f'Accuracy: {classifier.score(X_test, y_test):.4f}')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = rf_classifier.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
```

```python
from sklearn.model_selection import cross_val_score

classifier = rf_classifier

scores = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=5)
scores
```
